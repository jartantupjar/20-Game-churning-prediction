## Churning Solution

```python
import pandas as pd
import numpy as np
import json
from IPython.display import display
file='bulkmatches.json'
```

with open(file) as f:
    dict_data=json.load(f)
dict_data

```python
c_data=pd.read_json(file,orient='index')
new_json=c_data.to_json(orient='records',date_unit='s')
new_output=json.loads(new_json)
data = pd.io.json.json_normalize(new_output)
data.reset_index(inplace=True)
data.head(3)
```

```python
c_data.head(3)
```

```python
player_data = pd.io.json.json_normalize(new_output,record_path='players',meta=['date','duration','mapId'],record_prefix='player_')
player_data.head(3)
```

```python
data.shape[0]
```

```python
result = pd.merge(data, player_data, how='inner', on=['date','duration','mapId'])
#keep values that do not belong to a team
#result = pd.merge(result, team_data, how='left', on=['date','duration','mapId','player_team'])
result
```

```python
result['date']=result['date'].astype('object')
result['date'] =pd.to_datetime(result['date'],unit='s',infer_datetime_format=True)
result
```

```python
col_list=['duration','finished','group','mapId','official','server','timeLimit','player_auth','player_degree','player_flair','player_points','player_score','player_team','team_score']
for column in col_list:
    print(column,result[column].unique())
```

```python
#Remove unecessary columns that logically have no significant value/ are duplicate
# Or those that I do not plan to use for this analysis due to its additional complexity
result.drop(['port','players','official','player_auth',
             'teams','player_events','player_flair','mapId','player_team'],axis=1,inplace=True)
result
```
