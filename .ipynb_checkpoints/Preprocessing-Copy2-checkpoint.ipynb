## Churning Solution

```python
import pandas as pd
import numpy as np
import json
from IPython.display import display
file='bulkmatches.json'
```

with open(file) as f:
    dict_data=json.load(f)
dict_data

```python
c_data=pd.read_json(file,orient='index')
new_json=c_data.to_json(orient='records',date_unit='s')
new_output=json.loads(new_json)
data = pd.io.json.json_normalize(new_output)
data.reset_index(inplace=True)
data.head(3)
```

```python
c_data.head(3)
```

```python
player_data = pd.io.json.json_normalize(new_output,record_path='players',meta=['date','duration','mapId'],record_prefix='player_')
player_data.head(3)
```

```python
team_data = pd.io.json.json_normalize(new_output,record_path='teams',meta=['date','duration','mapId'],record_prefix='team_')
team_data.head(100)
```

```python
player_data['player_team'].value_counts()
```

```python
team_data['team_name'].value_counts()
```

```python
#notable sequence to determining the correct team is using alternate values.
team_data['player_team'] = 1
team_data.loc[1::2,"player_team"] = 2
team_data[:10]
```

```python
from tqdm import tqdm
group_team_data=team_data.groupby(['date','duration','mapId'],as_index=False).sum()
team_data['win']=0
#print(group_team_data)
for index, row in tqdm(group_team_data.iterrows()):
    #print(x)
    date=row['date']
    duration=row['duration']
    mapId=row['mapId']
    
    tm1_filter=(team_data['date']==date)&(team_data['duration']==duration)&(team_data['mapId']==mapId)&(team_data['player_team']==1)
    tm2_filter=(team_data['date']==date)&(team_data['duration']==duration)&(team_data['mapId']==mapId)&(team_data['player_team']==2)
    team_1=team_data.loc[tm1_filter]
    team_2=team_data.loc[tm2_filter]
    #print(row['date'])
    if not (team_1.empty):
        if not (team_2.empty):
            tm1=team_1['team_score'].values[0]
            tm2=team_2['team_score'].values[0]

            if tm1> tm2: 
                team_data.loc[tm1_filter,'win']=1
            elif tm1< tm2: 
                team_data.loc[tm2_filter,'win']=1
team_data
```

```python
#group team score together/ this is to avoid duplicating each of our players data
#this also removes team name and team splats as it doesnt seem important for churning
#team_data=team_data.groupby(['date','duration','mapId'],as_index=False).sum()
```

```python
result = pd.merge(data, player_data, how='inner', on=['date','duration','mapId'])
#keep values that do not belong to a team
result = pd.merge(result, team_data, how='inner', on=['date','duration','mapId','player_team'])
result
```

```python
print(result['team_score'].isnull().value_counts())
print(result['win'].isnull().value_counts())
#fill teamscore with zeroes for players that entered early.
result['team_score'].fillna(value=0,inplace=True)
result['win'].fillna(value=0,inplace=True)
```

```python
result['date']=result['date'].astype('object')
result['date'] =pd.to_datetime(result['date'],unit='s',infer_datetime_format=True)
result
```
